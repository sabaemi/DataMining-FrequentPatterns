Reading Dataset

```python
df = pd.read_csv(r'divar_posts_dataset.csv')
df1 = pd.read_csv(r'orders.csv')
```

Extracting the patterns of product sales posts in different cities in Divar dataset

```python
item = df.groupby('city')
trans_city = []
for item, i in item:
    trans_city.append(tuple(i['cat1']))

itemsets, rules = apriori(trans_city, min_support=1, min_confidence=1)
print(itemsets)
print(len(itemsets))
print(rules)
print(len(rules))
```

Extraction the sales patterns of different products in the Digikala dataset

```python
item=df1.groupby('ID_Customer')
trans_cust=[]
for item,i in item:
    trans_cust.append(tuple(i['ID_Item']))
import pyfpgrowth
itemsets = pyfpgrowth.find_frequent_patterns(trans_cust,4)
rules = pyfpgrowth.generate_association_rules(itemsets, 0.25)

print(itemsets)
print(len(itemsets))
print(rules)
print(len(rules))
```

Extraction the frequent patterns between the publication day of the post and the used platform

```python
split_data = df["created_at"].str.split(" ")
data = split_data.to_list()
names = ["created_at", "hour"]
new_df = pd.DataFrame(data, columns=names)
products_list = new_df.values.tolist()
df['created_at']=new_df['created_at']

trans_city=[]
item=df.groupby('id')
for item,i in item:
  trans_city.append(tuple(i['created_at']))
trans=[]
item=df.groupby('id')
for item,i in item:
    trans.append(tuple(i['platform']))

c = [trans_city[i]+trans[i] for i in range(len(trans_city))]

from efficient_apriori import apriori
itemsets, rules = apriori(c, min_support=0.13, min_confidence=0.2)
print(itemsets)
print(len(itemsets))
print(rules)
print(len(rules))
```

Extracting the frequent patterns for the products and the used platform

```python
df['cat2']=df['cat2'].fillna(method='bfill')
trans_city=[]
item=df.groupby('id')
for item,i in item:
  trans_city.append(tuple(i['cat2']))

trans=[]
item=df.groupby('id')
for item,i in item:
    trans.append(tuple(i['platform']))

c = [trans_city[i]+trans[i] for i in range(len(trans_city))]

from efficient_apriori import apriori
itemsets, rules = apriori(c, min_support=0.01, min_confidence=0.4)
print(itemsets)
print(len(itemsets))
print(rules)
print(len(rules))
```
